```python
# Drop not necessary columns
df_cleaned = df.drop(['Id','EmployeeName','Notes','Status','Agency'], axis=1)
df_cleaned.sample(5)
```

```python
# Calculate the mean of the non-missing values in the 'Benefits' column
benefits_mean = df_cleaned['Benefits'].mean()

# Replace missing values in the 'Benefits' column with the calculated mean
df_cleaned['Benefits'].fillna(benefits_mean, inplace=True)
```

```python
df_cleaned.shape
```

```python
# Missing values in each column
df_cleaned.isnull().sum()
```

```python
# Data Cleaning
# Dropping rows with missing values in any column, we have a big dataset
df_cleaned = df_cleaned.dropna()
df_cleaned.shape
```

We need to check the negative values in our dataset!

```python
# Filter the DataFrame to include only numerical columns
numerical_columns =['BasePay','OvertimePay', 'OtherPay', 'Benefits','TotalPay', 'TotalPayBenefits' ]
# Count the number of rows where any numerical column contains a negative value
num_negative_rows = len(df_cleaned[(df_cleaned[numerical_columns] < 0).any(axis=1)])

print("Number of negative salaries:", num_negative_rows)
```

```python
# Delete the negative rows
df_cleaned = df_cleaned.loc[(df_cleaned[numerical_columns] >= 0).all(axis=1)]
```

```python
df_cleaned.describe()
```

```python
df_cleaned.info()
```

```python
# Basic Data Visualization
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(df_cleaned['TotalPay'], bins=15, color='red', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
import re
df_cleaned['Department'] = df_cleaned['JobTitle'].str.extract(r'\(([A-z]?.*?)\)', expand=False).fillna('Not found')
```

```python
df_cleaned.Department.unique()
```

```python
# Define a mapping for similar department names
department_mapping = {
    'POLICE DEPARTMENT': 'Police Department',
    'FIRE DEPARTMENT': 'Fire Department',
    'CIVIL & CRIMINAL': 'Civil & Criminal',
    'CIVIL/CRIMINAL': 'Civil & Criminal',
    'MFCC': 'Not found',
    'SEASONAL': 'Seasonal',
    'Fire Dept': 'Fire Department',
    'Cvl&Crmnl': 'Civil & Criminal',
    'SFERS': 'Not found',
    'Civil/Criminal': 'Civil & Criminal'
}

# Map department names to a standard format
standardized_departments = [department_mapping.get(dept, dept) for dept in df_cleaned.Department]

# Convert the standardized department names back to an array
df_cleaned['departments_array'] = standardized_departments

```

```python
df_cleaned['departments_array'].unique()
```

```python
plt.figure(figsize=(8, 8))
# Filter out rows where 'Department' is not 'Not found' and then count occurrences
dept_counts = df_cleaned[df_cleaned['departments_array'] != 'Not found']['departments_array'].value_counts()

dept_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Employees in Different Departments')
plt.ylabel('')
plt.show()
```

The average salary for the Fire Department is relatively high compared to other departments, with a significant range between the minimum and maximum salaries.


```python
# Grouped Analysis
department_stats = df_cleaned.groupby('departments_array')['TotalPay'].agg(['mean', 'median', 'min', 'max'])
department_stats.sort_values(by='mean', ascending=False)
```

```python
# Simple Correlation Analysis
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_cleaned, y='Benefits', x='BasePay')
plt.title('Scatter Plot of Base Pay vs Total Pay')
plt.xlabel('Base Pay')
plt.ylabel('Benefits')
plt.grid(True)
plt.show()
```

Insights:
- The Fire Department has the highest average salary, followed by the Police Department and Civil & Criminal department.
- Seasonal positions have the lowest average salary, indicating part-time or temporary roles.
- The "Not found" category may require further investigation due to its wide range of salaries.
- OvertimePay and OtherPay have wide ranges, indicating significant variations in compensation.
- There seems to be some correlation between Base Pay and Benefits for the employees, as shown in the scatter plot.
